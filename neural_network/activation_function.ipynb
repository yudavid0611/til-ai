# 활성화 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 계단 함수

```python
# 계단 함수
def step_function(x:np.array):
    y = x > 0
    y.astype(int)
    return y
```

```python
# -5.0. -4.9 ... 4.9, 5.0
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 시그모이드 함수

```python
# 시그모이드 함수
def sigmoid(x):
    y = 1 / (1 + np.exp(-x))
    return y
```

```python
# -5.0. -4.9 ... 4.9, 5.0
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## ReLU 함수

```python
def relu(x):
    y = np.maximum(0, x)
    return y
```

```python
# -5.0. -4.9 ... 4.9, 5.0
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
# 그래프 한 번에 보기
y_step = step_function(x)
y_sigmoid = sigmoid(x)
y_relu = relu(x)

f, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].plot(x, y_step)
axes[0].set_ylim(-0.1, 1.1)
axes[0].set_title('step function')

axes[1].plot(x, y_sigmoid)
axes[1].set_ylim(-0.1, 1.1)
axes[1].set_title('sigmoid')

axes[2].plot(x, y_relu)
axes[2].set_ylim(-0.1, 1.1)
axes[2].set_title('ReLU')

plt.show()
```
